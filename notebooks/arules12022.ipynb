```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

### Example used in the presentation

```python
data1 = [['a1','a3','a4'],
           ['a2', 'a3', 'a5'],
           ['a1', 'a2', 'a3', 'a5'],
           ['a2', 'a5']]
#print(type(data1))
```

```python
te = TransactionEncoder()
te_ary = te.fit(data1).transform(data1)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)

frequent_itemsets
```

```python
#Finding the association rules with a confidence greater or equal that 0.8
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
```

### Example taken from the mlxtend homepage

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)

frequent_itemsets
```

```python
#Finding the association rules with a confidence greater or equal that 0.7
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules
```

```python
rules[rules['antecedents'] == {'Eggs', 'Kidney Beans'}]
```

## Example: Market basket of 7501 transactions and no more of 20 items in each of them

```python
#the data is available at https://stackabuse.com/association-rule-mining-via-apriori-algorithm-in-python/
store_data = pd.read_csv('https://academic.uprm.edu/eacuna/store_data.csv',header=None)  
store_data.head()
```

```python
#converting the dataframe in a list
records = []  
for i in range(0, 7501):  
    records.append([str(store_data.values[i,j]) for j in range(0, 20)])
print(type(records))
```

```python
te = TransactionEncoder()
te_ary = te.fit(records).transform(records)
df = pd.DataFrame(te_ary, columns=te.columns_)
df1=df.drop("nan",axis=1)
```

```python
frequent_itemsets = apriori(df1, min_support=0.05, use_colnames=True)

frequent_itemsets
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
```
