### Imputacion media/mediana/moda: Census Dataset

```python
import pandas as pd
datos = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', 
    header=None, 
    sep=',',na_values=[' ?'])

datos.columns=['v1', 'v2', 'v3', 'v4', 'v5','v6','v7','v8','v9','v10','v11', 'v12','v13','v14','class']
print(datos.shape)
```

```python
datos.head()
```

```python
#Excluyendo la columna de clase: class
x=datos.iloc[:,:-1]
```

```python
x.head(40)
```

```python
#Imputando por la moda
dfclean = x.apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
#Imputacion por la moda usando scikit-learn si la variable es cuantitativa y por la moda si la variable es  cualitativa 
import numpy as np
from sklearn.preprocessing import Imputer
from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)
X=x
xt = DataFrameImputer().fit_transform(X)
```

```python
#Imputacion por la mediana usando scikit-learn si la variable es cuantitativa y por la moda si la variable es  cualitativa 
import numpy as np
from sklearn.preprocessing import Imputer
from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].median() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)
X=x
xt = DataFrameImputer().fit_transform(X)
```
