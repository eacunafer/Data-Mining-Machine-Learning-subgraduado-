### Visualization of Missing Values

In this notebook will be using the library missingno created by Aleksey Bilogur https://github.com/ResidentMario The missigno module allow us visualization of missing values. Dataset: Census
Dataset: Census

```python
import numpy as np
import missingno as msno
#from matplotlib import gridspec
#import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#Leyendo la base de datos census de la internet
df = pd.read_csv('http://academic.uprm.edu/eacuna/census.csv',  sep=',',na_values=[' ?'])
df.columns
```

```python
msno.bar(df,color="Green")
```

### Displaying an image of the dataframe showing where the missing values are located

```python
msno.matrix(df,color=(.2,.4,.6),sparkline=False)
```
